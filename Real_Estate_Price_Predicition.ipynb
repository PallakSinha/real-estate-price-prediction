```python
false_count = (~df['IsRERARegistered']).sum()
false_count
```

```python
df['RERAID'] = df['RERAID'].fillna('NA')
```

```python
df['RERAID'].isnull().sum()
```

```python
null_counts_per_row = df.isnull().sum(axis=1)
threshold = 3
records_with_many_nulls = df[null_counts_per_row >= threshold]
records_with_many_nulls
```

```python
df.drop(records_with_many_nulls.index, inplace=True)
```

```python
df.shape
```

```python
df_grouped= df.groupby(['PropertyType', 'BHK'])
```

```python
df['Bathrooms']=df_grouped['Bathrooms'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Bathrooms'].isnull().sum()
```

```python
from datetime import date
```

```python
current_year = date.today().year

df.loc[df['YearBuilt'].isnull(), 'YearBuilt'] = current_year - df['AgeYears']
```
